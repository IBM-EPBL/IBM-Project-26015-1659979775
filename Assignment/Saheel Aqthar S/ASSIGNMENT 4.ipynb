```python
# importing necessary librariries
import pandas as pd
import numpy as np
import sklearn as sk
```

```python
from sklearn.preprocessing import scale
```

```python
names = data.columns
```

```python
data = scale(data)
```

```python
data
```

```python
data=pd.DataFrame(data,columns=names)
```

```python
data.head()
```

```python
from sklearn.cluster import KMeans
```

```python
x = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i,init = 'k-means++')
  kmeans.fit(data)
  x.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1,11),x)
plt.show()
```

```python
model = KMeans(n_clusters=2)
```

```python
pred = model.fit_predict(data)
```

```python
pred
```

```python
pred = pd.Series(pred)
```

```python
pred
```

```python
data['pred'] = pred
```
